```python
"""BeautifulSoup in build on the top of html5lib , lxml , html.parser , etc so
BeautifulSoup object and specify the parser library can be created at the same time
"""


import csv

URL = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_area"
r = requests.get(URL)

soup = BeautifulSoup(r.content, 'html5lib')

countries=[] # a list to store quotes

table = soup.find('div', attrs = {'class':'wikitable sortable sticky-header col2left jquery-tablesorter'})

for row in table.findAll('tr'):
	country = {}
	country['Name'] = row.td.text
	countries.append(country)

filename = 'test.csv'
with open(filename, 'w', newline='') as f:
	w = csv.DictWriter(f,['Name'])
	w.writeheader()
	for country in countries:
		w.writerow(country)
```
